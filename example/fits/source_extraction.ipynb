# Source Extraction

As MYRaf is primarily designed for astronomical photometric analysis, source extraction is a fundamental feature. To facilitate this, the `Fits` class provides two essential methods:

## daofind
extracts sources.


```python
from myraflib import Fits

fits = Fits.sample()

sources = fits.daofind(sigma=1.2)
fits.show(sources=sources)
```

## extract
extracts sources.

```python
sources = fits.extract()
fits.show(sources=sources)
```
