# Photometry

Photometry is the primary purpose of MYRaf. The software provides two methods for performing photometric analysis:

```python
from myraflib import Fits

fits = Fits.sample()
sources = fits.extract()
sources
```

## photometry_sep

This method performs photometry using the `sum_circle` function.

```python
photometry_set = fits.photometry_sep(sources["xcentroid"].to_list(), sources["ycentroid"].to_list(), 10)
photometry_set
```

## photometry_phu

This method performs photometry using the `CircularAperture` function.

```python
photometry_phu = fits.photometry_phu(sources["xcentroid"].to_list(), sources["ycentroid"].to_list(), 10)
photometry_phu
```

## photometry

Or one can use both methods at once:

```python
photometry = fits.photometry(sources["xcentroid"].to_list(), sources["ycentroid"].to_list(), 10)
photometry
```
