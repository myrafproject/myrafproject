```python
import warnings
from astropy.utils.exceptions import AstropyWarning

warnings.filterwarnings('ignore', category=AstropyWarning)
```

```python
from myraflib import Fits

fits = Fits.sample()
fits.data()
```

```python
sources = fits.extract()
sources
```

# photometry_sep
A `Fits` object can do photometry using `sum_circle`

```python
photometry_set = fits.photometry_sep(sources["xcentroid"].to_list(), sources["ycentroid"].to_list(), 10)
photometry_set
```

# photometry_phu
A `Fits` object can do photometry using `CircularAperture`

```python
photometry_phu = fits.photometry_phu(sources["xcentroid"].to_list(), sources["ycentroid"].to_list(), 10)
photometry_phu
```

# photometry
Or using both methods

```python
photometry = fits.photometry(sources["xcentroid"].to_list(), sources["ycentroid"].to_list(), 10)
photometry
```
