```python
import math
from myraflib import Fits

fits = Fits.sample()
fits.show()
```

# Shift
A `Fits` object can shift the image

```python
shifted_fits = fits.shift(100, 100)
shifted_fits.show()
```

# Rotate
A `Fits` object can rotate the image

```python
rotated_fits = fits.rotate(math.radians(45))
rotated_fits.show()
```

# Crop
A `Fits` object can crop the image

```python
cropped_fits = fits.crop(10, 10, 100, 100)
cropped_fits.show()
```

# Bin
A `Fits` object can bin the image

```python
binned_fits = fits.bin(5)
binned_fits.show()
```

# Align

A `Fits` object can align the image

```python
to_be_aligned = fits.shift(5, 5).rotate(math.radians(2))
to_be_aligned.show()
```

```python
aligned = to_be_aligned.align(fits)
aligned.show()
```
