```python
from myraflib import Fits

fits = Fits.sample()
fits
```

# Get Header
## header
A `Fits` object can obtain header from a fits file

```python
headers = fits.header()
headers
```

## pure_header
`pure_header` returns the header as `astropy` `Header`

```python
pure_header = fits.pure_header()
pure_header
```

# Alter Header

## hedit
`hedit` can create/update/delete header of a fits object

### create
if a none existing card is specified a new header card will be created

```python
fits.hedit("Test", "MYRaf", "Comment") # Here Test is card, MYRaf is value abd Comment is a comment
fits.pure_header()
```

### update
if an existing card is specified the header will be updated

```python
fits.hedit("Test", "MYRaf V3", "Updated Via MYRaf") # Here Test is card, MYRaf is value and Comment is a comment
fits.pure_header()
```

You can update an existing or creata a new header with value of another card

```python
fits.hedit("Test", "BITPIX", "From BITPIX", value_is_key=True) # Here Test is card, BITPIX is cark which its value would be used and Comment is a comment
fits.pure_header()
```

## Delete
if parameter is true the header card would be deleted no matter other parameters

```python
fits.hedit("Test", delete=True) # Here Test is card
fits.pure_header()
```
