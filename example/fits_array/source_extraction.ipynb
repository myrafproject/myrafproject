# Source Extraction

As MYRaf is primarily designed for astronomical photometric analysis, source extraction is a fundamental feature. To facilitate this, the `FitsArray` class provides two essential methods:

## daofind
extracts sources.


```python
from myraflib import FitsArray

fits_array = FitsArray.sample()

sources = fits_array.daofind(sigma=1.2)
fits_array[0].show(sources=sources)
```

## extract
extracts sources.

```python
sources = fits_array.extract()
fits_array[0].show(sources=sources)
```
