# show

The `FitsArray` object also provides functionality to visualize the data it contains.

```python
from myraflib import FitsArray

fits_array = FitsArray.sample()
fits_array.show()
```

The `show` method offers two optional parameters: `scale` and `interval`. The `scale` parameter is a boolean, and when set to `True`, the data will be z-scaled before visualization. The `interval` parameter is an integer and is the amination interval value.


```python
fits = fits_array[0]
fits.show()
```
