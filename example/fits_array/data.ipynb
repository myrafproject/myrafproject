# Data

Simular to `Fits`, `FitsArray` object does not provide a method to directly alter the data. However, it offers several methods to retrieve the data.

```python
from myraflib import FitsArray

fits_array = FitsArray.sample()
fits_array
```

## data

The `data` method returns the digital image data as a `numpy.ndarray`.

```python
fits_array.data()
```

## value

The `value` method returns the value of a specified pixel.

```python
fits_array.value(0, 0)
```

## ccd

The `ccd` method returns a `CCDData` object representing the specified FITS data.

```python
fits_array.ccd()
```
