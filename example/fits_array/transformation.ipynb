# Transformation

The `Fits` class offers a range of methods for transforming digital images. It strives to preserve the World Coordinate System (WCS) information, if present, throughout these transformations. So does the `FitsArray`...

```python
from myraflib import FitsArray

fits_array= FitsArray.sample()
fits_array.show()
```

## shift

The `FitsArray` class allows you to shift a digital images by a specified amount. This functionality is useful for aligning images where rotation is not a concern.

```python
shifted_fits_array = fits_array.shift(100, 100)
shifted_fits_array.show()
```

## Rotate

The `FitsArray` class provides a method to rotate the digital images by a specified angle, given in radians.

```python
from math import radians

rotated_fits = fits_array.rotate(radians(45))
rotated_fits.show()
```

## Crop

The `FitsArray` class includes a method for cropping the images. This method returns the specified subwindow of the image, defined by the lowest x and y coordinates, width, and height. Parameters required are:

- **x-coordinate** of the lowest corner
- **y-coordinate** of the lowest corner
- **width** of the subwindow
- **height** of the subwindow


```python
cropped_fits = fits_array.crop(100, 100, 500, 500)
cropped_fits.show()
```

## Bin

The `FitsArray` class features a `bin` method for binning the digital images. Binning can be useful for various photometric techniques, where combining adjacent pixels into larger bins helps in data analysis and noise reduction.

```python
binned_fits = fits_array.bin(10)
binned_fits.show()
```

## Align

The `FitsArray` class includes a method for all `Fits` object within it. This functionality helps in registering images to ensure that corresponding features are accurately matched, which is crucial for comparative analysis and combining data.

```python
aligned_fits = fits_array.align()
aligned_fits.show()
```
