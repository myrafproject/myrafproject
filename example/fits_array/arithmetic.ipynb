```python
import warnings
from astropy.utils.exceptions import AstropyWarning

warnings.filterwarnings('ignore', category=AstropyWarning)
```

```python
from myraflib import FitsArray

fits_array = FitsArray.sample()
fits_array.data()
```

# add
A `FitsArray` object can add values to fits list. The value may be another filts list, fits or a numeric value

## Numeric

```python
added_fits_array = fits_array.add(20)
added_fits_array.data()
```

## Fits

## FitsArray

```python
added_fits_array = fits_array.add(added_fits_array)
added_fits_array.data()
```

```python
added_fits_array = fits_array.add(added_fits_array[0])
added_fits_array.data()
```
