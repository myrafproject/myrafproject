# Arithmetic Operations

The `FitsArray` object offers a range of methods for performing arithmetic operations. In certain cases, it overrides standard operators to enhance the usability and flexibility of these operations. Just as `Fits` does.

## add

The `add` method allows you to add a value to the `FitsArray` object. The value can be either a numeric value, a `Fits` object, a list of values, another `FitsArray` or even a list of `Fits`. This method facilitates straightforward arithmetic operations with the `Fits` data.

```python
from myraflib import FitsArray

fits_array = FitsArray.sample(verbose=True)
```

```python
fits_array.show()
```

```python
added_fits = fits_array.add(20)
(added_fits - fits_array).imstat()
```
